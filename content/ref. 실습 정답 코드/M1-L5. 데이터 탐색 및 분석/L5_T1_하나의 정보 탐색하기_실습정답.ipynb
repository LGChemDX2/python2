## **(2) 실습1 : 수치형 정보 분석**

* ① 데이터 셋에 .describe()를 이용하여 기초 통계량을 조회합니다.
* ② total_bill에 대해 히스토그램, kde plot, box plot을 그리고 비즈니스를 파악해 봅시다.
* ③ tip에 대해 히스토그램, kde plot, box plot을 그리고 비즈니스를 파악해 봅시다.

① 데이터 셋에 .describe()를 이용하여 기초 통계량을 조회합니다.

```python
data.describe()
```

② total_bill에 대해 히스토그램, kde plot, box plot을 그리고 비즈니스를 파악해 봅시다.

```python
plt.figure(figsize = (12,4))

# 히스토그램
plt.subplot(1,3,1 )
plt.hist(data['total_bill'], bins = 30, edgecolor = 'gray')

# kde plot
plt.subplot(1,3,2 )
sns.kdeplot(data['total_bill'])

# box plot
plt.subplot(1,3,3 )
plt.boxplot(data['total_bill' ], vert = False)

plt.tight_layout()
plt.show()
```

* 비즈니스를 파악해 봅시다.

```python
# 밀집 : 10 ~ 20달러 정도 지불
# 희박 : 30 이상 
```

③ tip에 대해 히스토그램, kde plot, box plot을 그리고 비즈니스를 파악해 봅시다.

```python
plt.figure(figsize = (12,4))

var = 'tip'

# 히스토그램
plt.subplot(1,3,1 )
plt.hist(data[var], bins = 30, edgecolor = 'gray')

# kde plot
plt.subplot(1,3,2 )
sns.kdeplot(data[var])

# box plot
plt.subplot(1,3,3 )
plt.boxplot(data[var], vert = False)

plt.tight_layout()
plt.show()
```

* 비즈니스를 파악해 봅시다.

```python
# 밀집 : 2 ~ 4
# 희박 : 6 ~ 
```

## **(3) 실습2 : 범주형 정보 분석**
다음 두 정보에 대한 범주형 변수의 기초  통계량과 막대그래프, 파이 차트를 통해 비즈니스를 파악해 봅시다.
* ① gender
* ② time

① gender

```python
# 기초 통계량1
data['gender' ].value_counts()
```

```python
# 기초 통계량2
data['gender'].value_counts(normalize = True)
```

```python
# 시각화
plt.figure(figsize = (8, 4))
plt.subplot(1,2,1 )
sns.countplot(x = 'gender', data = data)

plt.subplot(1,2, 2)
temp = data['gender' ].value_counts().sort_index()
plt.pie(temp.values, labels = temp.index, autopct = '%.2f%%')

plt.tight_layout()
plt.show()
```

* 비즈니스를 파악해 봅시다.

```python
# 남자 : 64%
# 여자 : 36%
```

② time

```python
# 기초 통계량1
var = 'time'
data[var].value_counts()
```

```python
# 기초 통계량2
data[var].value_counts(normalize= True)
```

```python
# 시각화
plt.figure(figsize = (8, 4))
plt.subplot(1,2,1 )
sns.countplot(x = var, data = data)

plt.subplot(1,2,2 )
temp = data[var].value_counts().sort_index()
plt.pie(temp.values, labels = temp.index, autopct = '%.2f%%')

plt.tight_layout()
plt.show()
```

* 비즈니스를 파악해 봅시다.

```python
# 디너 : 72%
# 런치 : 28%
```
