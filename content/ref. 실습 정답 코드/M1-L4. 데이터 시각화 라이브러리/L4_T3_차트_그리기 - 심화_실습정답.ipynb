```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

* 데이터 불러오기

```python
kospi = pd.read_csv('KOSPI_2018.csv')
kospi['Date'] = pd.to_datetime(kospi['Date'])
kospi2 = kospi.loc[kospi['Date'].between('2018-01-01', '2018-01-31')]
kospi2.head()
```

```python
plt.figure(figsize = (8,4))
plt.plot('Date', 'Close', data = kospi2)
plt.grid()
plt.show()
```

## **(2) 실습1 : 겹쳐서 그리기**

* Open(시가)과 Close(종가)를 겹쳐서 그려 봅시다.
    * x축: Date, y축: Open, Close
    * 레이블 표시
    * marker 지정 : 's'
    * x축 값 회전 : 45

```python
plt.figure(figsize = (8,4))
plt.plot( 'Date', 'Open', data = kospi2, label = 'Open', marker = 's')
plt.plot( 'Date', 'Close', data = kospi2, label = 'Close', marker = 's')

plt.xticks(rotation= 45)
plt.legend()
plt.grid()
plt.show()
```

## **(3) 실습2 : 분할 프레임**

* 다음 조건에 맞게 그래프를 그려봅시다.
    * 프레임 : 1행 2열
    * 전체 크기 : 12, 4
    * 그래프1
        * x축 : Date, y 축 : Close
        * 선 색 : blue
        * marker : 'o'
        * grid 추가
        * title : Close
        * x축 값 회전 : 45
    * 그래프2
        * x축 : Date, y 축 : Volume
        * 선 색 : darkred
        * marker : 'o'
        * grid 추가
        * title : Volume
        * x축 값 회전 : 45

```python
# 전체 크기
plt.figure(figsize = (12, 4))

# 그래프1
plt.subplot(1,2,1)
plt.plot('Date', 'Close', data = kospi2, marker = 'o', color= 'blue')
plt.title('Close')
plt.xticks(rotation =45)
plt.grid()

# 그래프2
plt.subplot(1,2,2)
plt.plot('Date', 'Volume', data = kospi2, marker = 'o', color = 'darkred')
plt.title('Volume')
plt.xticks(rotation = 45)
plt.grid()

plt.tight_layout() # 프레임의 여백을 적절히 조정해주기
plt.show()
```
